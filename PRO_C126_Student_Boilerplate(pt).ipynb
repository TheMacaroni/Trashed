## Obtenha a Ação

```python
def get_action(current_state, reward_matrix):
    available_action = []
    print("reward_matrix","\n",reward_matrix)    
    for action in enumerate(reward_matrix[current_state]):     
        if action[1]!= -1:            
            available_action.append(action[0]) 
    choose_action = random.choice(available_action)
    print("Estado Atual",current_state)
    print("A escolha aleatória da ação de",available_action,"é", choose_action)          
    return choose_action
```

##Matriz Q
O **Aprendizado Q** é um algoritmo de aprendizado por reforço. Dado o estado atual, ele ajuda a encontrar a melhor ação a ser tomada pelo agente. No Aprendizado Q, o Q significa Qualidade. A qualidade representa quanto a ação é útil para obter uma recompensa.
Para realizar o Aprendizado Q, usamos a **Matriz-Q**. Ela também possui a forma de estados como linhas e ações como colunas. Inicialmente, todos os elementos da matriz Q estão zerados.


```python
q_matrix = np.zeros([6,6])
print(q_matrix)
```

## Tome uma Ação

```python
def take_action(current_state, reward_matrix, gamma):
    
    action = get_action(current_state, reward_matrix)
    
     #Estado, ação e recompensa atuais
    current_sa_reward = reward_matrix[current_state, action]

    #Novos estado, ação e recompensa
    q_sa_value = max(q_matrix[action,])
  
    #Estado Q atual
    q_current_state = current_sa_reward + (gamma * q_sa_value)

    #Atualize a matriz Q
    q_matrix[current_state,action] = q_current_state
    
    print("Matriz Q ","\n", q_matrix)

    new_state = action
    
    print("********************************************************************")
    
    if new_state == 5:
        print("Objetivo Atingido!")
    else: 
        print(f"Estado Anterior:{current_state} Novo Estado: {new_state}")
        
    return new_state
```

## Treinar o Modelo RL: Matriz Q
O objetivo é fazer com que o agente explore diferentes estados e alcance o objetivo; as recompensas são atualizadas após cada tentativa.

O agente conclui um **episódio** quando começa do estado inicial e atinge o objetivo.

Para cada ação, a matriz Q é atualizada. Isso atuará como o ambiente treinado para o agente. Usando a matriz Q atualizada, o agente pode procurar a recompensa máxima e encontrar o caminho ideal para qualquer estado.

Consulte o link abaixo para verificar como os episódios são executados e a matriz Q é atualizada.

```python
gamma = 0.8

# Obtenha a tabela Q
q_table = train(20, rewards, gamma)

print("Tabela Q Final: ", q_table)
```
